```python
plt.xlabel('Year')
plt.ylabel('Shipments')
plt.title('Quarterly Shipments Totals')
yearLabels = ['1985', '1986', '1987', '1988', '1989']

plt.xticks(np.arange(len(yearLabels)), yearLabels,rotation=0)

Q1=[4009,4123,4493,4595,4245]
Q2=[4321,4522,4806,4799,4900]
Q3=[4224,4657,4551,4417,4585]
Q4=[3944,4030,4485,4258,4533]

plt.plot(Q1, color='purple', label='Q1')
plt.plot(Q2, color='green', label='Q2')
plt.plot(Q3, color='blue', label='Q3')
plt.plot(Q4, color='red', label='Q4')  

red=mpat.Patch(color='purple', label='Q1')  
blue=mpat.Patch(color='green', label='Q2')  
green=mpat.Patch(color='blue', label='Q3')  
purple=mpat.Patch(color='red', label='Q4')

plt.legend(handles=[red, blue, green, purple])
plt.show()
```

<h3>Question 3.3(a)

```python
df_laptop = pd.read_csv(DATA / "LaptopSalesJanuary2008.csv")
print(type(df_laptop))
```

```python
from numpy import mean
ax = sns.barplot(x="Store Postcode", y="Retail Price", color='b', data=df_laptop, estimator=mean)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
ax.set_xticklabels(ax.get_xticklabels(), rotation=80)
ax.set(ylim=(450, 520))
plt.title('Average Retail Prices per Store')
```

Based off of this bar chart, the highest is N17 6QA (494.6). The lowest average is W4 3PH (481). 

<h2>Question 3.3(b)

```python
axs = sns.boxplot(x="Store Postcode", y="Retail Price", color='b', data=df_laptop)
axs.set_xticklabels(axs.get_xticklabels(), rotation=80)
```

There does not seem to be a large discrepancy between the two stores. The highest average and lowest average stores are both around the same in terms of distribution of price. The only difference is that everything is slightly higher in N17 6QA than in W4 3PH.
